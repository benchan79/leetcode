```python
# Create a DataFrame from List
# My Solution
def createDataframe(student_data: List[List[int]]) -> pd.DataFrame:
  data_dict = {}
  idx = range(len(student_data))
  for i in idx:
    data_dict[i+1] = student_data[i]
  df = pd.DataFrame(data_dict, index=['student_id', 'age'])
  df = df.transpose()
  return df

sd = [
  [1, 15],
  [2, 11],
  [3, 11],
  [4, 20]
]

df = createDataframe(sd)

df
# Suggested Solution
def createDataframe(student_data: List[List[int]]) -> pd.DataFrame:
  column_names = ['student_id', 'age']
  df = pd.DataFrame(student_data, columns=column_names)
  return df

sd = [
  [1, 15],
  [2, 11],
  [3, 11],
  [4, 20]
]

df = createDataframe(sd)

df
```

```python
# Reshape Data: Pivot
import random

def pivotTable(weather: pd.DataFrame) -> pd.DataFrame:
  pivot_df = weather.pivot(index='month', columns='city', values='temperature')
  month_order = ["Jan", "Feb", "Mar", "Apr", "May",]
  pivot_df = pivot_df.reindex(month_order)
  # pivot_df.reset_index(inplace=True)
  # pivot_df.columns.name = None
  return pivot_df

city_list = ['Jacksonville']*5 + ['ElPaso']*5
mth_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May']*2
tmp_list = [random.randint(1, 40) for _ in range(10)]

data = {
    'city': city_list,
    'month': mth_list,
    'temperature': tmp_list,
}

df = pd.DataFrame(data)

ss = pivotTable(df)
ss
```

id_vars: This specifies the columns that should remain unchanged. For this problem, only the product column remains unchanged because we want every row in the output to be associated with a product.

value_vars: This specifies the columns that we want to "melt" or reshape into rows. In our case, these are the sales data columns for each quarter: quarter_1, quarter_2, quarter_3, and quarter_4.

var_name: This is the name of the new column that will store the header names from the value_vars. In our problem, these are the quarter names.

value_name: This is the name of the new column that will store the values from the value_vars. In our problem, this will be the sales figures for each product for each quarter.

```python
# Reshape Data: Melt

def meltTable(report: pd.DataFrame) -> pd.DataFrame:
  report = pd.melt(report, id_vars=['product'], var_name='quarter', value_name='sales')
  return report

# Suggested
# report = report.melt(
#     id_vars=["product"],
#     value_vars=["quarter_1", "quarter_2", "quarter_3", "quarter_4"],
#     var_name="quarter",
#     value_name="sales",
# )

data = {
    'product': ['Umbrella', 'SleepingBag'],
    'quarter_1': [417, 800],
    'quarter_2': [224, 936],
    'quarter_3': [379, 93],
    'quarter_4': [611, 875],
}

df = pd.DataFrame(data)

ss = meltTable(df)
ss
```
