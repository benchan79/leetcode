```python
#  Customers Who Never Order

def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
  order_set = set(orders['customerId'].to_list())
  cust_set = set(customers['id'].to_list())
  cust_no_order = list(cust_set - order_set)
  id_no_order = customers['id'].isin(cust_no_order)
  subset = customers.loc[id_no_order, ['name']]
  subset.rename(columns={'name': 'Customers'}, inplace=True)
  return subset

def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
  merged_df = pd.merge(
    left=customers,
    right=orders,
    left_on="id",
    right_on='customerId',
    how='left',
  )
  no_order = merged_df['customerId'].isna()
  subset = merged_df.loc[no_order, ['name']]
  return subset.rename(columns={'name': 'Customers'})


data1 = {
  'id': list(range(1, 5)),
  'name': ['Joe', 'Henry', 'Sam', 'Max']
}

data2 = {
  'id': [1, 2],
  'customerId': [3, 1],
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

ss = find_customers(df1, df2)
ss
```

```python
# Nth Highest Salary
import pandas as pd
def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:
  subset = employee.drop_duplicates(subset=['salary'])
  subset = subset.sort_values(by=['salary'], ascending=False, ignore_index=True)
  subset.rename(columns={'salary': f'getNthHighestSalary({N})'}, inplace=True)
  subset.drop(columns={'id'}, inplace=True)
  if N <= subset.shape[0] and N > 0:
    subset = subset.iloc[N-1:N,:]
    return subset
  else:
    return pd.DataFrame({'getNthHighestSalary({})'.format(N): [None]})


data = {
  'id': [1, 2, 3],
  'salary': [100, 200, 300]

}
df = pd.DataFrame(data)
ss = nth_highest_salary(df, 2)
ss
```

```python
# Second Highest Salary

def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:
  subset = employee.rename(columns={'salary': 'SecondHighestSalary'})
  subset = subset.drop_duplicates(subset=['SecondHighestSalary'])
  subset = subset.nlargest(2, columns='SecondHighestSalary')
  if len(subset) >= 2:
    subset = subset.iloc[-1:,-1:]
    return subset
  else:
    return pd.DataFrame({'SecondHighestSalary': [None]})

data = {
  'id': [1,2,3,4],
  'salary':[300,200,200,100,]
}
df = pd.DataFrame(data)
ss = second_highest_salary(df)
ss
```

```python
#  Rank Scores

def order_scores(scores: pd.DataFrame) -> pd.DataFrame:
  uniq_scores = pd.DataFrame({'score': sorted(scores['score'].unique()), 'rank': range(scores['score'].nunique(), 0, -1)})
  final = pd.merge(
    left= scores['score'],
    right=uniq_scores[['score', 'rank']],
    on='score',
  )
  final.sort_values(by=["rank"], ascending=True, inplace=True)
  return final

data = {
  'id': [1,2,3,4,5,6],
  'score': [3.50, 3.65, 4.00, 3.85, 4.00, 3.65],
}

df = pd.DataFrame(data)
ss = order_scores(df)
# ss

# uniq_scores_df = pd.DataFrame({'score': sorted(df['score'].unique()), 'rank': range(df['score'].nunique(), 0, -1)})
# final = pd.merge(
#   left= df['score'],
#   right=uniq_scores_df[['score', 'rank']],
#   on='score',
# )
# final.sort_values(by=["rank"], ascending=True, inplace=True)
# final
```

```python
#  Rearrange Products Table

def rearrange_products_table(products: pd.DataFrame) -> pd.DataFrame:
  products = products.melt(
    id_vars=['product_id'],
    value_vars=['store1', 'store2', 'store3'],
    var_name='store',
    value_name='price'
  )
  products.dropna(subset=['price'], inplace=True)
  products.sort_values(by=['product_id', 'store'], inplace=True)
  return products

data = {
  'product_id': [0,1],
  'store1': [95, 70],
  'store2': [100, None],
  'store3': [105, 80]
}

df = pd.DataFrame(data)
ss = rearrange_products_table(df)
ss

# df = df.melt(
#   id_vars=['product_id'],
#   value_vars=['store1', 'store2', 'store3'],
#   var_name='store',
#   value_name='price'
# )
# df.dropna(subset=['price'], inplace=True)
# df.sort_values(by=['product_id', 'store'], inplace=True)
# df
```

```python
# Find Total Time Spent by Each Employee

def total_time(employees: pd.DataFrame) -> pd.DataFrame:
    employees['total_time'] = employees['out_time'] - employees['in_time']
    employees = employees.groupby(['event_day', 'emp_id'])[['total_time']].sum()
    employees.reset_index(inplace=True)
    employees.rename(columns={'event_day': 'day'}, inplace=True)
    return employees

data = {
  'emp_id' : [1,1,1,2,2],
  'event_day' : [
        "2020-11-28",
        "2020-11-28",
        "2020-12-03",
        "2020-11-28",
        "2020-12-09",
  ],
    'in_time': [4,55,1,3,47],
    'out_time': [32,200,42,33,74]
}

df = pd.DataFrame(data)
ss = total_time(df)
ss
# df['total_time'] = df['out_time'] - df['in_time']
# df = df.groupby(['event_day', 'emp_id'])[['total_time']].sum()
# df.reset_index(inplace=True)
# df.rename(columns={'event_day': 'day'}, inplace=True)
# df
```

```python
# Replace Employee ID With The Unique Identifier

def replace_employee_id(employees: pd.DataFrame, employee_uni: pd.DataFrame) -> pd.DataFrame:
  merged = pd.merge(
    left=employees,
    right=employee_uni,
    how='left',
    on='id'
  )
  merged = merged.loc[:, ['unique_id', 'name']]
  return merged

data1 = {
  'id': [1,7,11,90,3],
  'name': ['Alice', 'Bob', 'Meir', 'Winston', 'Jonathan']
}

data2 = {
  'id': [3,11,90],
  'unique_id': [1,2,3]
}

employees = pd.DataFrame(data1)
employee_uni = pd.DataFrame(data2)
ss = replace_employee_id(employees, employee_uni)
ss
# merged = pd.merge(
#   left=employees,
#   right=employee_uni,
#   how='left',
#   on='id'
# )
# merged = merged.loc[:, ['unique_id', 'name']]
# merged
```

```python
# Students and Examinations

def students_and_examinations(students: pd.DataFrame, subjects: pd.DataFrame, examinations: pd.DataFrame) -> pd.DataFrame:
  merged = pd.merge(
    left=pd.merge(students, subjects, how='cross'),
    right=examinations,
    how='left',
    on=['student_id'],
    suffixes=("", "_exam")
  )
  filt = merged['subject_name_exam'] == merged['subject_name']
  merged['attended_exams'] = filt.astype(int)
  merged.drop(columns=['subject_name_exam'], inplace=True)
  merged = merged.groupby(['student_id', 'student_name', 'subject_name'], dropna=False).sum().reset_index()
  merged.sort_values(by=['student_id', 'subject_name'], inplace=True)
  return merged

# Students table
students_data = {
    'student_id': [1, 2, 13, 6],
    'student_name': [None, 'Bob', 'John', 'Alex']
}

students = pd.DataFrame(students_data)

# Subjects table
subjects_data = {
    'subject_name': ['Math', 'Physics', 'Programming']
}

subjects = pd.DataFrame(subjects_data)

# Examinations table
examinations_data = {
    'student_id': [1, 1, 1, 2, 1, 1, 13, 13, 13, 2, 1],
    'subject_name': ['Math', 'Physics', 'Programming', 'Programming', 'Physics', 'Math',
                     'Math', 'Programming', 'Physics', 'Math', 'Math']
}

examinations = pd.DataFrame(examinations_data)

merged = pd.merge(
  left=pd.merge(students, subjects, how='cross'),
  right=examinations,
  how='left',
  on=['student_id'],
  suffixes=("", "_exam")
)

filt = merged['subject_name_exam'] == merged['subject_name']
merged['attended_exams'] = filt.astype(int)
merged.drop(columns=['subject_name_exam'], inplace=True)
merged = merged.groupby(['student_id', 'student_name', 'subject_name'], dropna=False).sum().reset_index()

merged.sort_values(by=['student_id', 'subject_name'], inplace=True)
merged


# ss = students_and_examinations(students, subjects, examinations)
# ss
```

```python
# Managers with at Least 5 Direct Reports

def find_managers(employee: pd.DataFrame) -> pd.DataFrame:
  mgr = employee.groupby('managerId').size().reset_index()
  mgr = mgr.loc[mgr[0] >= 5, ['managerId']]
  mgr = pd.merge(
    left=mgr,
    right=employee[['name', 'id']],
    how='inner',
    left_on='managerId',
    right_on='id',
  )
  mgr = mgr.loc[:, ['name']]
  return mgr

data = {
    'id': [101, 102, 103, 104, 105, 106],
    'name': ['John', 'Dan', 'James', 'Amy', 'Anne', 'Ron'],
    'department': ['A', 'A', 'A', 'A', 'A', 'B'],
    'managerId': [None, 101, 101, 101, 101, 101]
    # 'managerId': [None, 100, 100, 100, 100, 100]
}

employee = pd.DataFrame(data)

mgr = employee.groupby('managerId').size().reset_index()
mgr = mgr.loc[mgr[0] >= 5, ['managerId']]
mgr = pd.merge(
  left=mgr,
  right=employee[['name', 'id']],
  how='inner',
  left_on='managerId',
  right_on='id',
)
mgr = mgr.loc[:, ['name']]
mgr
```
